```python
f,  (ax1, ax2,ax3,ax4,ax5,ax6,ax7)  =  plt.subplots(nrows=7, ncols=1, figsize=(10, 30))
#1
ax1 = d_total[['Year','Rape']].groupby(['Year']).sum().plot(kind='bar',
                                       legend = True, title ="Rape victims", 
                                       color='Green', fontsize=12, ax=ax1)
ax1.set_xlabel("Year", fontsize=12)
ax1.set_ylabel("Rape victims", fontsize=12)

#2
ax2 = d_total[['Year','Dowry Deaths']].groupby(['Year']).sum().plot(kind='bar',
                                       legend = True, title ="Dowry victims", 
                                       color='Red', fontsize=12,ax=ax2)
ax2.set_xlabel("Year", fontsize=12)
ax2.set_ylabel("Dowry Deaths", fontsize=12)

#3
ax3 = d_total[['Year','Kidnapping and Abduction']].groupby(['Year']).sum().plot(kind='bar',
                                       legend = True, title ="Kidnapping and Abduction victims", 
                                       color='Orange', fontsize=12,ax=ax3)
ax3.set_xlabel("Year", fontsize=12)
ax3.set_ylabel("Kidnapping and Abduction", fontsize=12)

#4
ax4 = d_total[['Year','Assault on women with intent to outrage her modesty']].groupby(['Year']).sum().plot(kind='bar',
                                       legend =True, title ="victims", 
                                       color='violet', fontsize=12,ax=ax4)
ax4.set_xlabel("Year", fontsize=12)
ax4.set_ylabel("assault on women", fontsize=12)



#5
ax5 = d_total[['Year','Insult to modesty of Women']].groupby(['Year']).sum().plot(kind='bar',
                                       legend = True, title =" Insult to modesty victims", 
                                       color='Red', fontsize=12,ax=ax5)
ax5.set_xlabel("Year", fontsize=12)
ax5.set_ylabel("Insult to modesty of Women ", fontsize=12)

#6
ax6 = d_total[['Year','Cruelty by Husband or his Relatives']].groupby(['Year']).sum().plot(kind='bar',
                                       legend = True, title =" husbend cruelity victims", 
                                       color='Green', fontsize=12,ax=ax6)
ax6.set_xlabel("Year", fontsize=12)
ax6.set_ylabel("Cruelty by Husband or his Relatives", fontsize=12)

#7
ax7 = d_total[['Year','Importation of Girls']].groupby(['Year']).sum().plot(kind='bar',
                                       legend = True, title ="Importation of Girls", 
                                       color='Green', fontsize=12,ax=ax7)
ax7.set_xlabel("Year", fontsize=12)
ax7.set_ylabel("Importation of Girls", fontsize=12)
f.tight_layout()
```
